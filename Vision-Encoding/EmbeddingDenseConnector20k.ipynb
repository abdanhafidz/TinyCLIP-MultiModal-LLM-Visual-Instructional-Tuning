```python
import torch
from PIL import Image
from transformers import CLIPProcessor, CLIPModel, CLIPVisionModel
from tqdm import tqdm
import os
import numpy as np

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load TinyCLIP model and processor
model_path = "wkcn/TinyCLIP-ViT-40M-32-Text-19M-LAION400M"
model =  CLIPVisionModel.from_pretrained(model_path)
processor = CLIPProcessor.from_pretrained(model_path)

# Move model to appropriate device
model.to(device)
model.eval()
```

```python
from einops import rearrange

def get_embedding(image_path):
    try:
        # Load image
        image = Image.open(image_path).convert('RGB')

        # Process image using CLIP processor
        inputs = processor(images=image, return_tensors="pt")
        inputs = inputs.to(device)

        # Get visual features before the final pooling
        with torch.no_grad():
            x = inputs['pixel_values']

            # Create attention masks (all True for vision)
            batch_size = x.shape[0]
            sequence_length = 50  # 49 patches + 1 CLS token
            attention_mask = torch.ones((batch_size, sequence_length), device=device)
            causal_attention_mask = torch.zeros((batch_size, sequence_length, sequence_length), device=device)

            # Get the patch embeddings
            embeddings = model.vision_model.embeddings(x)

            # Pass through encoder layers and collect multi-layer features
            hidden_states = model.vision_model.pre_layrnorm(embeddings)

            # Collect features from multiple layers
            layer_features = []
            total_layers = len(model.vision_model.encoder.layers)
            # Extract from 4 layers: 1/4, 1/2, 3/4, and last layer
            # extract_at = [total_layers//4, total_layers//2, 3*total_layers//4, total_layers-1]

            for layer_idx, layer in enumerate(model.vision_model.encoder.layers):
                hidden_states = layer(hidden_states,
                                   attention_mask=attention_mask,
                                   causal_attention_mask=causal_attention_mask)[0]

                # Extract features at specified layers
                if layer_idx in range(total_layers):
                    # Remove CLS token
                    features = hidden_states[:, 1:, :]  # Remove CLS token
                    H = W = int(np.sqrt(features.shape[1]))  # Should be 7x7 for 224x224 input

                    # Reshape to spatial format (B, C, H, W)
                    spatial_features = rearrange(features, 'b (h w) c -> b c h w', h=H, w=W)

                    # # Global average pooling to get vector representation [B, C]
                    # pooled_features = spatial_features.mean(dim=[2, 3]).squeeze(0)  # [C]

                    layer_features.append(spatial_features)

            # Return format expected by Dense Connector: [[layer1, layer2, ...]]
            return [layer_features]

    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return [[]]  # Return empty list format that Dense Connector can handle
```

```python
from datasets import Dataset

dataset = Dataset.from_parquet('/content/dataset.parquet')
```

```python
def update_conversation(conversation):
    for turn in conversation:
        turn['role'] = turn.pop('from')  # Replace 'from' with 'role'
        if turn['role'] == 'human':
            turn['role'] = 'user'
        elif turn['role'] == 'gpt':
            turn['role'] = 'assistant'
    return conversation

dataset = dataset.map(lambda x: {'conversations': update_conversation(x['conversations'])}, batched=False)
```
